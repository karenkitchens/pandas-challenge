## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
female = purchase_data.loc[purchase_data["Gender"] == "Female"]
male = purchase_data.loc[purchase_data["Gender"] == "Male"]
non = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]

male_count = male["SN"].nunique()
female_count = female["SN"].nunique()
non_count = non["SN"].nunique()
total = purchase_data["SN"].nunique()

male_percentage = round(male_count / total * 100, 2)
female_percentage = round(female_count / total * 100, 2)
non_percentage = round(non_count / total * 100, 2)
gender_dict = [
    {"Gender": "Male", "Count": male_count, "Percentage": male_percentage}, 
    {"Gender": "Female", "Count": female_count, "Percentage": female_percentage}, 
    {"Gender": "Other / Non-Disclosed", "Count": non_count, "Percentage": non_percentage},   
             ]
gender_analysis = pd.DataFrame(gender_dict)
gender_analysis
```

## Age Demographics

```python
label_bins = ['< 10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data['age_bins'] = pd.cut(purchase_data["Age"], bins=[0, 9, 14, 19, 24, 29, 34, 39, 150], labels=label_bins)
purchase_data
```

```python
age_demos = pd.DataFrame(purchase_data)
less_than_ten = age_demos["age_bins"] == "< 10"
ten_to_fourteen = age_demos["age_bins"] == "10-14"
fifteen_to_nineteen = age_demos["age_bins"] == "15-19"
twenty_to_twenty_four = age_demos["age_bins"] == "20-24"
twenty_five_to_twenty_nine = age_demos["age_bins"] == "25-29"
thirty_to_thirty_four = age_demos["age_bins"] == "30-34"
thirty_five_to_thirty_nine = age_demos["age_bins"] == "35-39"
over_fourty = age_demos["age_bins"] == "40+"

count_all = age_demos["SN"].nunique()
count_less_than_ten = age_demos[less_than_ten]["SN"].nunique()
count_ten_to_fourteen = age_demos[ten_to_fourteen]["SN"].nunique()
count_fifteen_to_nineteen = age_demos[fifteen_to_nineteen]["SN"].nunique()
count_twenty_to_twenty_four = age_demos[twenty_to_twenty_four]["SN"].nunique()
count_twenty_five_to_twenty_nine = age_demos[twenty_five_to_twenty_nine]["SN"].nunique()
count_thirty_to_thirty_four = age_demos[thirty_to_thirty_four]["SN"].nunique()
count_thirty_five_to_thirty_nine = age_demos[thirty_five_to_thirty_nine]["SN"].nunique()
count_over_fourty = age_demos[over_fourty]["SN"].nunique()

age_df = pd.DataFrame([
    {"Age": "<10", "Total Count": count_less_than_ten, "Percentage": round(count_less_than_ten / count_all * 100, 2)},
    {"Age": "10-14", "Total Count": count_ten_to_fourteen, "Percentage": round(count_ten_to_fourteen / count_all * 100, 2)},
    {"Age": "14-19", "Total Count": count_fifteen_to_nineteen, "Percentage": round(count_fifteen_to_nineteen / count_all * 100, 2)},
    {"Age": "20-24", "Total Count": count_twenty_to_twenty_four, "Percentage": round(count_twenty_to_twenty_four / count_all * 100, 2)},
    {"Age": "25-29", "Total Count": count_twenty_five_to_twenty_nine, "Percentage": round(count_twenty_five_to_twenty_nine / count_all * 100, 2)},
    {"Age": "30-34", "Total Count": count_thirty_to_thirty_four, "Percentage": round(count_thirty_to_thirty_four / count_all * 100, 2)},
    {"Age": "35-39", "Total Count": count_thirty_five_to_thirty_nine, "Percentage": round(count_thirty_five_to_thirty_nine / count_all * 100, 2)},
    {"Age": "40+", "Total Count": count_over_fourty, "Percentage": round(count_over_fourty / count_all * 100, 2)},
])

age_df.set_index("Age")
```
